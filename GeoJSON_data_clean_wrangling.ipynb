```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
path_shp = 'immi/immi_shapefile.shp'
immi = gpd.GeoDataFrame.from_file(path_shp)
```

```python
immi.head()
```

```python
ls = ['Field'+str(i) for i in range(21, 29)]
```

```python

immi['2010'] = immi['Field21'] + immi['Field22'] + immi['Field23'] +\
            immi['Field24'] + immi['Field25'] + immi['Field26'] + immi['Field27'] + immi['Field28']
```

```python
immi = immi.drop(columns=['Field21',
 'Field22',
 'Field23',
 'Field24',
 'Field25',
 'Field26',
 'Field27',
 'Field28'])
```

```python
immi = immi.rename(columns={
    "Field1" : "countryName",
    "Field2" : 1820,
    "Field3" : 1830,
    "Field4" : 1840,
    "Field5" : 1850,
    "Field6" : 1860,
    "Field7" : 1870,
    "Field8" : 1880,
    "Field9" : 1890,
    "Field10" : 1900,
    "Field11" : 1910,
    "Field12" : 1920,
    "Field13" : 1930,
    "Field14" : 1940,
    "Field15" : 1950,
    "Field16" : 1960,
    "Field17" : 1970,
    "Field18" : 1980,
    "Field19" : 1990,
    "Field20" : 2000,
})
```

```python
ls = [x for x in range (1820, 2010,10)]
for i in ls:
    immi[i] = immi[i].str.replace(',', '').astype(int)
```

```python
immi.head()
```

```python
immi_new = immi.drop(list(immi)[0:13], axis=1)
immi_new 
```

```python
immi_new = immi_new.drop(list(immi_new)[-4:-1], axis=1)
immi_new.head()
```

```python
immi_new_melt = pd.melt(immi_new, id_vars=['countryName'], var_name="Decade", value_name = "Immi_num")
```
